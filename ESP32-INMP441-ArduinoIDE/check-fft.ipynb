### IMPORT LIBRARY

```python
import numpy as np
from numpy import sin, cos, pi
from scipy.fftpack import fft, fftfreq, ifft
import matplotlib.pyplot as plt
import librosa
import IPython.display as ipd
import soundfile as sf
import pandas as pd
```

### IMPORT SOUND

```python
x = pd.read_csv("sound_samples_voicenoised.csv")
x = x.values.flatten()
x
```

### COMPUTE FAST FOURIER TRANSFORM

```python
#compute spectrum
sampling_rate = 10000
yhat=fft(x)
fcycles=fftfreq(len(x),1/sampling_rate) # Frekuensi dalam rad/s
# Fpositive=np.where(fcycles>=0)
```

### PLOT DATA IN FREQUENCY DOMAIN

```python
plt.plot(fcycles,np.absolute(yhat))
plt.xlabel('Frequency(Hz)',fontsize=14)
plt.ylabel('Amplitude Spectrum',fontsize=14)
plt.grid()
```
