### IMPORT LIBRARY

```python
import numpy as np
from numpy import sin, cos, pi
from scipy.fftpack import fft, fftfreq, ifft
import matplotlib.pyplot as plt
import librosa
import IPython.display as ipd
import soundfile as sf
import pandas as pd
```

### IMPORT SOUND

```python
x = pd.read_csv("sound_samples.csv")
x = x.values.flatten()
x
```

### COMPUTE FAST FOURIER TRANSFORM

```python
#compute spectrum
sampling_rate = 10000
spectrum=fft(x)
N = len(x)
F=fftfreq(N,1/sampling_rate)
Fpositive=np.where(F>=0)
```

### PLOT DATA IN FREQUENCY DOMAIN

```python
# plt.plot(F[Fpositive],np.absolute(spectrum[Fpositive])/N)
plt.plot(F[F>4000],np.absolute(spectrum[F>4000])/N)
# plt.plot(F,np.absolute(spectrum)/N)
plt.xlabel('Frequency(Hz)',fontsize=14)
plt.ylabel('Amplitude Spectrum',fontsize=14)
plt.grid()
```
