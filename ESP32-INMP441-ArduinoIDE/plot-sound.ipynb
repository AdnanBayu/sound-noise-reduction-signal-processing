### IMPORT LIBRARY

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("sound_samples_voicenoised.csv")
data
```

```python
samplingFreq = 10000 # sampled at 10 kHz frequency = 10000 sample/s
tlims = [0,10]       # time duration in second (s)
time = np.linspace(tlims[0],tlims[1],(tlims[1]-tlims[0])*samplingFreq)
```

```python
# plt.plot(time[2000:3000], data[2000:3000])
plt.plot(time[:len(data)], data)
plt.title("sound sample")
plt.xlabel("time")
plt.ylabel("magnitude")
```
