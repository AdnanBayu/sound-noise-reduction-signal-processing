```python
import numpy as np
from numpy import sin, cos, pi
from scipy.fftpack import fft, fftfreq, ifft
import matplotlib.pyplot as plt
import librosa
import IPython.display as ipd
import soundfile as sf
import pandas as pd
```

```python
x = pd.read_csv("sound_samples1.csv")
x = x.values.flatten()
x
```

```python
#compute spectrum
sampling_rate = 10000
spectrum=fft(x)
N = len(x)
F=fftfreq(N,1/sampling_rate)
Fpositive=np.where(F>=0)
```

```python
t = np.array(range(0, len(x))) / sampling_rate

plt.plot(F[Fpositive],np.absolute(spectrum[Fpositive])/N)
# plt.plot(F,np.absolute(spectrum)/N)
plt.xlabel('Frequency(Hz)',fontsize=14)
plt.ylabel('Amplitude Spectrum',fontsize=14)
plt.grid()
```
